```python
import cohere
from dotenv import load_dotenv
import os

# Load your Cohere API key from an environment variable
# load_dotenv()
# api_key = os.getenv("N7qG6ka2MQ3Tt7PcIheilFVKlWtRONo7UY7qX70n")

# Initialize the Cohere client
co = cohere.Client("N7qG6ka2MQ3Tt7PcIheilFVKlWtRONo7UY7qX70n")

# # Read the contents of your 10-K filings
# with open(r"C:\Users\srima\Fintech\amazon_10k_fillings\sec-edgar-filings\AMZN\10-K\0000891020-00-000622\full-submission.txt", "r", encoding="utf-8") as f:
#     text = f.read()


filings_dir = r"C:\Users\srima\Fintech\apple_7section"

# Initialize an empty string to store the concatenated text
combined_text = ""

# Loop through all files in the directory
for filename in os.listdir(filings_dir):
    file_path = os.path.join(filings_dir, filename)
    
    # Check if the file is a text file
    if file_path.endswith(".txt"):
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()
            combined_text += text + "\n\n" # Separate files with newlines


# Define a prompt to generate insights from the 10-K filings
prompt = f"Based on the following 10-K filings given, provide a detailed analysis and insights considering the its progression over time such that I have to make a visualization from the insighs, so give proper insights which I can make a visualization from them :\n\n{combined_text}\n\nAnalysis and Insights:"


# Use the Cohere API's generate function to generate the analysis and insights
response = co.generate(
    model='command-r-plus',
    prompt=prompt,
    max_tokens=5000,
    temperature=0.7,
    k=0,
    p=1,
    frequency_penalty=0,
    presence_penalty=0,
    stop_sequences=["--"],
    return_likelihoods='NONE')

# Print the generated analysis and insights
# print(response.generations[0].text)
```
